```python
import matplotlib.pyplot as plt

# Data
guesses = [4, 5, 6, 7, 8, 9, 10, 11, 12]
percentages = [0, 0.3, 11.6, 53.7, 84.2, 96.9, 99.4, 99.8, 100]

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(guesses, percentages, color='skyblue', edgecolor='black')

# Add text annotations above each bar
for bar, percentage in zip(bars, percentages):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{percentage}%', 
             ha='center', va='bottom', fontsize=10)

# Labeling
plt.title('Quordle Games Solved in n Guesses', fontsize=14, weight='bold')
plt.xlabel('Number of Guesses (n)', fontsize=12)
plt.ylabel('Percentage of Games Solved (%)', fontsize=12)
plt.xticks(guesses, fontsize=10)
plt.yticks(range(0, 101, 10), fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
